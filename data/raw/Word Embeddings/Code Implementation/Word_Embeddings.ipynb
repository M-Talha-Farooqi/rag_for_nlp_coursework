# 📚 Word2Vec Embeddings – Theoretical Overview

---

## 🔹 What is Word2Vec?

**Word2Vec** is a technique used to convert words into **dense vector representations**, also known as **word embeddings**.

These embeddings capture **semantic relationships** between words by placing similar words **closer together in a continuous vector space**.

For example:


---

## 🔹 Why Use Word Embeddings?

- Traditional methods (like one-hot encoding) treat each word as independent and unrelated.
- Word2Vec, however, captures **context and meaning**.
- It reduces **dimensionality** and preserves **semantic similarity**.

---

## 🔹 How Does Word2Vec Work?

Word2Vec is based on the idea that **a word is defined by the company it keeps** (distributional hypothesis).

There are two main architectures:

1. **CBOW (Continuous Bag of Words):**  
   - Predicts a word based on its **context words**.
   - Example: Predict `"attack"` from `["the", "on", "was"]`.

2. **Skip-Gram:**  
   - Predicts **context words** from a given word.
   - Example: Use `"attack"` to predict `"the"`, `"on"`, `"was"`.

Both methods use a shallow neural network to learn embeddings. The **hidden layer's weights become the word vectors**.

---

## 🔹 What Do the Embeddings Represent?

- Each word is represented by a **vector of real numbers** (e.g., 50 or 100 dimensions).
- These vectors encode:
  - **Semantic similarity** (e.g., `"terror"` close to `"violence"`)
  - **Syntactic roles**
  - **Analogical reasoning** (e.g., `"paris" - "france" + "italy" ≈ "rome"`)

---

## 🔹 Custom vs. Pretrained Word2Vec Embeddings

### ✅ Custom-Trained Embeddings

- You train Word2Vec on **your own dataset** (e.g., 50 terrorism-related sentences).
- Advantages:
  - Tailored to your domain and vocabulary.
  - Captures context specific to your data.
- Disadvantages:
  - Needs sufficient text data to be meaningful.
  - Poor generalization outside the training corpus.

### ✅ Pretrained Word2Vec Embeddings

- Trained on massive corpora like **Google News**, **Wikipedia**, or **Common Crawl**.
- Common sources:
  - `GoogleNews-vectors-negative300.bin` (3 million words, 300 dimensions)
- Advantages:
  - Very rich semantic information.
  - Ready to use and effective even with small datasets.
- Disadvantages:
  - May include vocabulary irrelevant to your task.
  - Large file sizes (can be hundreds of MBs).

---

## 🔹 Summary

| Feature           | Custom-Trained                | Pretrained                          |
|------------------|-------------------------------|-------------------------------------|
| Data Source       | Your own dataset              | Public corpora (Google News, etc.)  |
| Vocabulary        | Limited to your text          | Broad and extensive                 |
| Domain Fit        | Excellent for your domain     | General-purpose                     |
| Training Needed   | Yes                           | No                                  |
| Quality           | Varies (needs data)           | High                                |

---

Word2Vec is a powerful way to represent language numerically and has paved the way for many modern NLP techniques. Whether you use custom or pretrained embeddings depends on your data size, domain specificity, and task requirements.

## 🧠 Traditional Text Vectorization Techniques (Before Word2Vec)

Before deep learning-based embeddings like **Word2Vec**, text data was converted into numerical form using simpler, frequency-based techniques. These approaches did not capture semantic meaning but were widely used in early Natural Language Processing (NLP) tasks.

---

### 🔹 1. Bag of Words (BoW)

- Represents text as a **vector of word counts**.
- Ignores word order and context.
- The size of the vector equals the number of unique words in the corpus.
- Example:
  - Sentence: "Terrorists threatened security"
  - Vector: [1, 1, 1] → one count for each word.

**Pros:**
- Simple and fast.
- Easy to implement.

**Cons:**
- No information about word meaning or position.
- High dimensional and sparse.

---

### 🔹 2. TF-IDF (Term Frequency – Inverse Document Frequency)

- Improves BoW by reducing the weight of **common words** and increasing the importance of **rare but informative words**.
- Formula:
  - `TF-IDF = TF(w,d) * log(N / DF(w))`
- Results in better text representations than raw counts.

**Pros:**
- Reduces impact of frequently occurring but less informative words like "the", "is", etc.
- Still simple to compute.

**Cons:**
- Still ignores word order and semantics.
- Sparse representation.

---

### ❗ Limitations of BoW and TF-IDF

- **No semantic understanding** (e.g., "car" and "automobile" are unrelated numerically).
- **Context is lost** — can't distinguish different meanings of the same word.
- **High-dimensional vectors** lead to inefficiency and overfitting.

➡️ These limitations led to the development of **dense, semantic word embeddings** like **Word2Vec**, which map words to vectors in a way that captures their **meaning and context**.


```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
# ==========================
# Bag of Words (BoW)
# ==========================
bow_vectorizer = CountVectorizer()
bow_features = bow_vectorizer.fit_transform(df['Text'])

# Convert to DataFrame
bow_df = pd.DataFrame(bow_features.toarray(), columns=bow_vectorizer.get_feature_names_out())
print("🔹 Bag of Words Features:")
display(bow_df)

# ==========================
# TF-IDF
# ==========================
tfidf_vectorizer = TfidfVectorizer()
tfidf_features = tfidf_vectorizer.fit_transform(df['Text'])

# Convert to DataFrame
tfidf_df = pd.DataFrame(tfidf_features.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
print("🔹 TF-IDF Features:")
display(tfidf_df)
```

## ⚙️ Word2Vec Training Parameters – Theoretical Explanation

---

### 🔹 `vector_size`

- Specifies the number of dimensions in which each word is represented.
- Determines the **size of the word embedding vector** (e.g., 5, 100, 300).
- A higher size can capture more complex semantic relationships but may require more data and training time.

---

### 🔹 `window`

- Sets the **context window size** — the number of words before and after the target word to consider as context.
- For example, if `window=3`, then for the word `"attack"`, the model will consider up to 3 words on either side.
- Smaller values focus on **local context**, while larger values capture **broader context**.

---

### 🔹 `min_count`

- Ignores all words with total frequency **lower than this threshold**.
- Helps filter out rare words that may add noise or be irrelevant.
- Useful for focusing the model on more meaningful and frequently occurring words.

---

### 🔹 `workers`

- Sets the number of **CPU threads** used during training.
- This allows parallelization and speeds up training on large datasets.

---

### 🔹 `sg`

- Determines the **training algorithm**:
  - `sg = 1`: Use **Skip-Gram** — predicts context words based on the current word.
  - `sg = 0`: Use **CBOW (Continuous Bag of Words)** — predicts the current word based on its context.
- **Skip-Gram** is better for learning rare words and works well for smaller datasets.
- **CBOW** is faster and effective on larger corpora.

---

These parameters guide how Word2Vec learns the relationships between words and builds dense vector representations that capture **semantic meaning**, **context**, and **word similarity**.


```python
# Train Word2Vec
model = Word2Vec(sentences, vector_size=5, window=3, min_count=1, workers=2, sg=1)
```

## 🧾 Printing Word Embeddings – Theoretical Explanation

---

### 🔹 Vocabulary in Word2Vec

Once a Word2Vec model is trained, it creates a **vocabulary** — a list of all words that met the `min_count` threshold during training.

Each of these words is assigned a **dense vector** (embedding) that captures its **semantic and contextual meaning**.

---

### 🔹 Accessing Embeddings

For every word in the vocabulary:
- The model stores a **corresponding vector** in its internal structure.
- This vector is a **real-valued array** (e.g., 5, 100, or 300 dimensions depending on `vector_size`).
- These vectors are stored in an ordered list called the **embedding matrix**.

---

### 🔹 Purpose of Printing Embeddings

When you loop through all words in the vocabulary and print their embeddings:
- You are **examining the learned representations** for each word.
- This helps you understand:
  - How the model perceives each word.
  - Which words have similar embeddings (i.e., similar meanings).
  - The structure of the **semantic space** created by the model.

---

### 🔹 Example Output Interpretation

For each word:
- You see a vector like `[0.12, -0.45, 0.33, ..., 0.08]`.
- These values **do not have human-readable meaning individually**.
- But together, they represent the **position of the word in vector space**, where distance and direction relate to meaning.

---

### 🔹 Practical Uses

- Visualizing embeddings (e.g., using t-SNE or PCA).
- Measuring **similarity** between words using cosine similarity.
- Using embeddings as **input features** for downstream NLP tasks (e.g., classification, clustering).

---

By printing all embeddings, you're effectively viewing the **knowledge Word2Vec has learned** about your dataset’s vocabulary.


```python
# Print embeddings for all words in the vocabulary
for word in model.wv.index_to_key:
    print(f"Word: '{word}'")
    print(f"Embedding: {model.wv[word]}\n")
```

## 📊 Understanding Sentence Embeddings from Word2Vec

---

### 🔹 Context

Once a Word2Vec model is trained, every word in the vocabulary is represented by a vector of real numbers (e.g., 5 values if `vector_size=5`).

To generate a **sentence embedding**, we combine the vectors of all words in the sentence — most commonly by computing the **average of their embeddings**.

---

### 🧩 Step-by-Step Example

Let’s analyze the following two sentences and their embeddings:

#### 🔸 Sentence 1:
**"The attack caused widespread panic"**

### 🔸 Step 1: Word Embeddings
Words and their embeddings:

| Word         | Embedding (vector of 5 values) |
|--------------|---------------------------------|
| the          | `[-0.192071, 0.10014586, -0.17519172, -0.0878365, -0.000702]`  
| attack       | `[-0.03875482, 0.16154873, -0.11861791, 0.00090325, -0.09507468]`  
| caused       | `[-0.15023164, -0.01860085, 0.19076237, -0.14638333, -0.04667537]`  
| widespread   | `[-0.06810732, -0.01892803, 0.11537147, -0.15043275, -0.07872207]`  
| panic        | `[0.14623532, 0.10140524, 0.13515386, 0.01525731, 0.12701781]`  

➡️ To compute the sentence embedding:
- Add all corresponding dimensions across the 5 vectors.
- Divide by the number of words (5 in this case).

### 🔸 Step 2: Add Embeddings Dimension-wise

We sum each dimension **across all 5 vectors**:

| Dimension | Sum of Values |
|-----------|----------------|
| 1st dim   | `-0.192071 + (-0.03875482) + (-0.15023164) + (-0.06810732) + 0.14623532 = -0.30392946`  
| 2nd dim   | `0.10014586 + 0.16154873 + (-0.01860085) + (-0.01892803) + 0.10140524 = 0.32557095`  
| 3rd dim   | `-0.17519172 + (-0.11861791) + 0.19076237 + 0.11537147 + 0.13515386 = 0.14747807`  
| 4th dim   | `-0.0878365 + 0.00090325 + (-0.14638333) + (-0.15043275) + 0.01525731 = -0.36849202`  
| 5th dim   | `-0.000702 + (-0.09507468) + (-0.04667537) + (-0.07872207) + 0.12701781 = -0.09415631`  

---

### 🔸 Step 3: Divide Each Sum by the Number of Words (5)

This gives the **average embedding** (sentence vector):

| Dimension | Calculation | Result |
|-----------|-------------|--------|
| 1st dim   | `-0.30392946 / 5` | `-0.06058589`  
| 2nd dim   | `0.32557095 / 5`  | `0.06511419`  
| 3rd dim   | `0.14747807 / 5`  | `0.02949562`  
| 4th dim   | `-0.36849202 / 5` | `-0.07369840`  
| 5th dim   | `-0.09415631 / 5` | `-0.01883126`  


📌 Final sentence embedding (mean of 5 word vectors):

`[-0.060585897, 0.06511419, 0.029495616, -0.07369841, -0.018831262]`

---

#### 🔸 Sentence 2:
**"Terrorists threatened national security"**

### 🔸 Step 1: Word Embeddings
Words and their embeddings:

| Word          | Embedding (vector of 5 values) |
|---------------|---------------------------------|
| terrorists    | `[-0.03632035, 0.0575316, 0.01983747, -0.1657043, -0.18897636]`  
| threatened    | `[0.1476101, -0.03066943, -0.09073226, 0.13108103, -0.09720321]`  
| national      | `[-0.14233617, 0.12917745, 0.17945977, -0.10030856, -0.07526743]`  
| security      | `[-0.01072454, 0.00472863, 0.10206699, 0.18018547, -0.186059]`  

### 🔸 Step 2: Add Embeddings Dimension-wise

We compute the sum of each corresponding dimension across all 4 word vectors:

| Dimension | Sum of Values |
|-----------|----------------|
| 1st dim   | `-0.03632035 + 0.1476101 + (-0.14233617) + (-0.01072454) = -0.04177096`  
| 2nd dim   | `0.0575316 + (-0.03066943) + 0.12917745 + 0.00472863 = 0.16076825`  
| 3rd dim   | `0.01983747 + (-0.09073226) + 0.17945977 + 0.10206699 = 0.210632`  
| 4th dim   | `-0.1657043 + 0.13108103 + (-0.10030856) + 0.18018547 = 0.04525364`  
| 5th dim   | `-0.18897636 + (-0.09720321) + (-0.07526743) + (-0.186059) = -0.547506`  

---

### 🔸 Step 3: Divide Each Sum by the Number of Words (4)

We compute the average of each dimension:

| Dimension | Calculation        | Result         |
|-----------|--------------------|----------------|
| 1st dim   | `-0.04177096 / 4`  | `-0.01044274`  
| 2nd dim   | `0.16076825 / 4`   | `0.04019206`  
| 3rd dim   | `0.210632 / 4`     | `0.05265799`  
| 4th dim   | `0.04525364 / 4`   | `0.01131341`  
| 5th dim   | `-0.547506 / 4`    | `-0.1368765`  

➡️ Final sentence embedding is the **average of the 4 word vectors**:

📌 Final sentence embedding:

`[-0.010442743, 0.040192064, 0.05265799, 0.011313409, -0.1368765]`

---

### 🧠 What Does This Mean?

- The resulting sentence vector captures the **combined semantic meaning** of the sentence.
- While individual numbers don't have standalone meaning, the position of the vector in high-dimensional space can be compared with other vectors to measure **similarity**, **clustering**, or **semantic distance**.

---

### 📌 Summary

- Each word is represented as a 5-dimensional vector learned by Word2Vec.
- To compute the sentence embedding, we take the **average of all word vectors** in the sentence.
- This aggregated vector represents the **semantic meaning of the entire sentence**, allowing it to be used in further NLP tasks such as classification, clustering, or similarity analysis.


```python
# Function to create sentence embedding
def get_sentence_embedding(tokens):
    vectors = [model.wv[token] for token in tokens if token in model.wv]
    return np.mean(vectors, axis=0) if vectors else np.zeros(model.vector_size)

# Apply to the DataFrame
df['embedding'] = [get_sentence_embedding(tokens) for tokens in sentences]


# Show results
df
```

```python
import pandas as pd
import numpy as np
import gensim.downloader as api
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')

# 1. Load pretrained Google Word2Vec model (takes time)
model = api.load("word2vec-google-news-300")  # 300-dimension vectors

# 2. Example sentences
data = {'Text': [
    'The attack caused widespread panic',
    'Terrorists threatened national security'
]}
df = pd.DataFrame(data)

# 3. Tokenize each sentence
df['tokens'] = df['Text'].apply(lambda x: word_tokenize(x.lower()))
df
```

```python
# 4. Function to get word embeddings
def get_word_embeddings(tokens):
    return {word: model[word] for word in tokens if word in model}

# 5. Function to get sentence embedding (mean of word vectors)
def get_sentence_embedding(tokens):
    vectors = [model[word] for word in tokens if word in model]
    return np.mean(vectors, axis=0) if vectors else np.zeros(model.vector_size)

```

```python
# 6. Apply to get embeddings
df['word_embeddings'] = df['tokens'].apply(get_word_embeddings)
df['sentence_embedding'] = df['tokens'].apply(get_sentence_embedding)

# 7. View results
pd.set_option('display.max_colwidth', None)
df
```

# BERT


```python
from transformers import BertTokenizer, BertModel
import torch

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

text = "The terrorist planted explosives near the government building."
inputs = tokenizer(text, return_tensors='pt')
outputs = model(**inputs)

# Get the embeddings for [CLS] token
cls_embedding = outputs.last_hidden_state[0][0]
print("BERT [CLS] Embedding Shape:", cls_embedding.shape)
print("BERT [CLS] Embedding Vector:\n", cls_embedding.detach().numpy())
```
